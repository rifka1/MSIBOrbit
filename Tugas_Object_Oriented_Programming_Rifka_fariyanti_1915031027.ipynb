```python
class hero:
  """Bantuan Library Hero:

  NAME
    hero
  
  DESCRIPTION
    Library untuk bermain MOBA sederhana untuk Python
    =================================================
    Library hero mendukung pembuatan hero dan
    memfasilitasi pertarungan hero.

    Sumber : https://drive.google.com/file/d/1jsHSA0UrShh6rXSp5zQtnvlv9kLwMo97/view?usp=sharing

  CLASSES
    fighter
    mage
  
  FUNCTION
    get_info_hero() -> tuple
    memberikan_serangan() -> float
    menerima_serangan(float)
    recall()
    ultimate() -> float
  """

  nama = ''
  hp = 0
  mana = 0
  attack = 0
  defense = 0
  max_hp = 0
  max_mana = 0
    
  def get_info_hero(self):
    return self.nama, self.hp, self.mana, self.attack, self.defense
  
  def memberikan_serangan(self):
    print(f'{self.nama} memberikan serangan sebesar {self.attack}')
    return self.attack
  
  def menerima_serangan(self, serangan):
    if serangan > self.defense:
      print(f'{self.nama} menerima serangan sebesar {serangan - self.defense}')
      self.hp = self.hp - (serangan - self.defense) # rumus menerima serangan
      print(f'sisa HP {self.nama} = {self.hp}')
    if self.hp <= 0:                              # jika HP habis, maka tumbang
      print(f'{self.nama} tumbang dan respawn')
      self.recall()
  
  def recall(self):
    print(f'sisa HP = {self.hp}, sisa Mana = {self.mana}')
    self.hp = self.max_hp
    self.mana = self.max_mana
    print(f'mengisi ulang HP {self.nama} menjadi {self.hp}')
    print(f'mengisi ulang Mana {self.nama} menjadi {self.mana}')
  
  def ultimate(self):
    print(f'Hero mengeluarkan skill ultimate!')

class fighter(hero):
  def __init__(self, nama, hp, mana, attack, defense):
    self.nama = nama
    self.hp = hp
    self.mana = mana
    self.attack = attack
    self.defense = defense
    self.max_hp = hp
    self.max_mana = mana

  def ultimate(self):
    print(f'{self.nama} mengeluarkan skill ultimate!')  # ultimate tanpa mana
    return self.attack * 4                              # rumus damage ultimate

class mage(hero):
  def __init__(self, nama, hp, mana, attack, defense):
    self.nama = nama
    self.hp = hp
    self.mana = mana
    self.attack = attack
    self.defense = defense
    self.max_hp = hp
    self.max_mana = mana
    
  def ultimate(self):
    if self.mana-100 >= 0:                       # ultimate perlu 100 mana
      print(f'{self.nama} mengeluarkan skill ultimate')
      self.mana -= 100                           # ultimate mengurangi 100 mana
      return self.attack * 4                     # rumus damage ultimate
    else:
      print(f'{self.nama} kehabisan mana!')
      return 0
```

2. Buat sebuah program OOP dengan ketentuan:

  - berbeda dari contoh program yang ada di materi (ppt dan colab)
  - memiliki 1 parent class
  - memiliki setidaknya 2 child class

  Setelah program selesai, buatlah:
  - setidaknya 2 instance untuk setiap child class

```python
class bahan_pakaian:
  """

  NAME
    bahan pakaian

  DESCRIPTION
    Bahan pakaian dijual dalam pembuatan beberapa jenis pakaian yaitu kaos, jaket, kemeja dll.
    Yang mana masing-masing dari bahan tersebut memiliki karakteristik dan harga yang berbeda.

  CLASSES
    katun
    denim
    flanel

  FUNCTION
    jenis_bahan() = ''
    karakteristik() = ''
    jumlah() = int
    harga() = int
    harga_pesanan() = int
    get_harga()= int
    hitung_potongan() = int
    get_potongan() = int
    __init__()
    
  """

  harga = 0

  def harga_pesanan(self):
   return self.harga

  def get_harga(self):
    print(f"""{self.nama} terbuat dari {self.jenis_bahan} dan memiliki karakteristik {self.karakteristik} dibeli dengan jumlah {self.jumlah} seharga Rp. {self.harga}""")

  def get_potongan(self):
    print(f'khusus di bulan februari maka akan mendapatkan potongan harga sebesar Rp. {self.potongan}')

#child class 1 (katun)
class katun(bahan_pakaian):
  def __init__(self, nama, jenis_bahan, karakteristik, jumlah):
    self.nama = nama
    self.jenis_bahan = jenis_bahan
    self.karakteristik = karakteristik
    self.jumlah = jumlah

  def harga_pesanan(self):
    self.harga = self.jumlah*35000

  def hitung_potongan(self):
    self.potongan = 0.5*self.harga

#child class 2 (denim)
class denim(bahan_pakaian):
  def __init__(self, nama, jenis_bahan, karakteristik, jumlah):
    self.nama = nama
    self.jenis_bahan = jenis_bahan
    self.karakteristik = karakteristik
    self.jumlah = jumlah

  def harga_pesanan(self):
    self.harga = self.jumlah*40000

  def hitung_potongan(self):
    self.potongan = 0.5*self.harga

#child class 3 (flanel)
class flanel(bahan_pakaian):
  def __init__(self, nama, jenis_bahan, karakteristik, jumlah):
    self.nama = nama
    self.jenis_bahan = jenis_bahan
    self.karakteristik = karakteristik
    self.jumlah = jumlah

  def harga_pesanan(self):
    self.harga = self.jumlah*30000

  def hitung_potongan(self):
    self.potongan = 0.5*self.harga

bahan_pakaian1 = katun('Kaos', 'katun', 'tipis', 2)
bahan_pakaian1.harga_pesanan()
bahan_pakaian1.get_harga()
bahan_pakaian1.hitung_potongan()
bahan_pakaian1.get_potongan()
print('\n')
bahan_pakaian2 = denim('Jaket', 'denim', 'tebal dan lembut', 1)
bahan_pakaian2.harga_pesanan()
bahan_pakaian2.get_harga()
bahan_pakaian2.hitung_potongan()
bahan_pakaian2.get_potongan()
print('\n')
bahan_pakaian3 = flanel('Kemeja', 'flanel', 'lembut dan berwarna gelap', 3)
bahan_pakaian3.harga_pesanan()
bahan_pakaian3.get_harga()
bahan_pakaian3.hitung_potongan()
bahan_pakaian3.get_potongan()
```
